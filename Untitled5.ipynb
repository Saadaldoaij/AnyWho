```python
from google.colab import drive

drive.mount('/content/gdrive')  # Mount your Google Drive to the '/content/gdrive' directory

import os
# Create a folder in your Drive to save results (if needed)
os.makedirs('/content/gdrive/MyDrive/MyColabResults', exist_ok=True)  # Create 'MyColabResults' folder in MyDrive if it doesn't exist

# Save your data or results to the Drive folder
with open('/content/gdrive/MyDrive/MyColabResults/my_data.txt', 'w') as f:
    f.write('This is my data to save.')

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from simplet5 import SimpleT5
import pytorch_lightning as pl
from pytorch_lightning import Trainer
import matplotlib.pyplot as plt
import re
from textblob import TextBlob
from spellchecker import SpellChecker
import string
```

```python
# Load the dataset
path = '/content/gdrive/MyDrive/MyColabResults/ReportsDATASET.csv'
df = pd.read_csv(path, usecols=['Text'])
```

```python
spell = SpellChecker()

def correct_spelling(word):
    return spell.correction(word) or word

def preprocess_text(text):
    # Noise Removal
    text = re.sub(r'\n', ' ', text)


    # Spell Check
    words = text.split()
    corrected_words = [correct_spelling(word) for word in words]
    text = ' '.join(corrected_words)

    # Tokenization & Stopword Removal
    words = [word for word in text.split() if word.lower() not in ENGLISH_STOP_WORDS]
    return ' '.join(words)

# Vectorized function
def vectorized_preprocessing(df, column_name):
    df['source_text'] = df[column_name].apply(preprocess_text)
    df['target_text'] = df['source_text']
    return df

```

```python
import re

def remove_placeholders(text):
    # Remove "xxx" or "XXXX" placeholders
 return re.sub(r'\b[Xx/]{1,}\b', '', text)


```

```python
df['source_text'] = df['Text'].apply(remove_placeholders)
df['target_text'] = df['Text'].apply(remove_placeholders)
```

```python
df['source_text'] =  df['Text'].str.replace('\n', ' ')
df['target_text'] =  df['Text'].str.replace('\n', ' ')

```

```python
import re

def remove_placeholders(text):
    # Regular expression to match various patterns of "X"
    return re.sub(r'\b[Xx/]{1,}\b', '', text)
df['source_text'] = df['source_text'].apply(remove_placeholders)
df['target_text'] = df['target_text'].apply(remove_placeholders)
```

```python

# Prepare data for T5 summarization
df['source_text'] = "Summarize: " + df['source_text']
train_df, test_df = train_test_split(df, test_size=0.3)
train_df.to_csv("/content/gdrive/MyDrive/MyColabResults/train.csv", index=False)
test_df.to_csv("/content/gdrive/MyDrive/MyColabResults/test.csv", index=False)
```

```python

# Initialize T5 model and train
model = SimpleT5()
model.from_pretrained(model_type="t5", model_name="t5-base")

model.train(train_df=train_df[:1388],
           eval_df=test_df[:596],
           source_max_token_len=600,
           target_max_token_len=200,
           batch_size=2, max_epochs=6, use_gpu=True)
```
